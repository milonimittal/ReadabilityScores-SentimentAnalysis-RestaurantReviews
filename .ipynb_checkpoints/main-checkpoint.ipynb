```python
#converting lower case and removing punctuations
reviews.text = reviews.text.str.lower()
reviews['removed_punct_text'] = reviews.text.str.replace('\n','').str.replace('[!"#$%&\()*+,-./:;<=>?@[\\]^_`{|}~]','')
```

```python
#loading easy words list
import csv
easy_words_open = open('../AllFiles/DaleChallEasyWordsList.txt')
reader =csv.reader(easy_words_open)
easy_words = [word[0] for word in reader]
easy_words = [element.lower() for element in easy_words]
```

```python
#counting number of words
def func(reviewx):
    words=reviewx.split()
    words_count = len(words)
    return words_count
reviews["words"] = reviews.removed_punct_text.apply(func)
reviews.head(5)
```

```python
#counting number of sentences
reviews["sentences"] = reviews.text.str.count("." or "?" or "!")
reviews.head(5)
```

```python
#finding number of easy words
def func2(reviewx):
    words = len([word for word in reviewx.split() if word in easy_words])
    return words
reviews["easy_words"] = reviews.removed_punct_text.apply(func2)
reviews.head(5)
```

```python
#finding the Dale Chall's index
reviews["DaleChall_index"] = (0.1579*(100*(reviews.words-reviews.easy_words))/(reviews.words))+(0.0496*(reviews.words/reviews.sentences)+3.6365)
reviews.head(5)
```

```python
#finding if usefulness depends on Dale Chall's index
stval = reviews.groupby('stars').mean()
stval.corr()
#correlation between Dale Chall index and number of people finding it useful is negative which means higher the index the lesser useful it is 
```

```python
### Sentiment analysis for positive or negative review

reviews['pos_neg']=["pos" if x>=3 else "neg" for x in reviews['stars']]
reviews.head(5)
```

```python
## import positive file which contains common meaningless positive words such as good
file_positive = open('../AllFiles/positive.txt')
reader =csv.reader(file_positive)
positive_words = [word[0] for word in reader]

## import negative file which contains common meaningless positive words such as bad
file_negative = open('../AllFiles/negative.txt')
reader =csv.reader(file_negative)
negative_words = [word[0] for word in reader]
```

```python
# making separate table of useful information
table=reviews[['removed_punct_text','pos_neg']]
```

```python
#keeping only the words which are in positive or negative words list
def filter_words(review):
    words = [word for word in review.split() if word in positive_words + negative_words]
    words = ' '.join(words)
    return words
table.removed_punct_text = table.removed_punct_text.apply(filter_words)
```

```python
#vectorizing the words
x = table['removed_punct_text']
y = table['pos_neg']
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x=vectorizer.fit_transform(x)
```

```python
#splitting into testing and training data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
#applying SVM
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm.fit(x_train, y_train)
#finding the predicted values of test reviews
y_pred_svm = svm.predict(x_test)
from sklearn.metrics import accuracy_score
print("SVM Accuracy Score -> ",accuracy_score(y_pred_svm, y_test)*100)
```

```python
#applying Bayes model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train, y_train)
y_predict_naive = nb.predict(x_test)
print("Naive Bayes Accuracy Score -> ",accuracy_score(y_predict_naive, y_test)*100)
```
